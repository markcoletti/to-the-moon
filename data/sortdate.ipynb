```python
import pandas as pd
import time
import numpy as np
```

```python
events = pd.read_csv('events.csv',parse_dates=True,infer_datetime_format=True)
```

```python
    daily_events = events[events.Date != 'Daily']
```

```python
daily_events = daily_events.replace('TBD',np.NaN)
```

```python
daily_events['Start Time'] = daily_events['Start Time'].map(lambda x: time.strptime(x,'%H:%M %p') if not pd.isna(x) else np.NaN)
```

```python
daily_events.sort_values('Start Time')
```

```python
days = ['Thursday','Friday','Saturday','Sunday']
```

```python
daily_events.Date = daily_events.Date.astype('category',ordered=True,categories=days)
```

```python
daily_events.sort_values(['Date','Start Time'])
```
